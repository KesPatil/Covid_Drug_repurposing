#### PCA based K-means clustering

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
#Apply Standard Scaling
sc = StandardScaler()
X_drug = sc.fit_transform(X_drug)
pd.DataFrame(X_drug)
```

```python
#Define two components
pca=PCA(n_components=50)
principalComponents=pca.fit_transform(X_drug)

#principalComponents
data = np.hstack((X_drug_name.reshape((322,1)),principalComponents))
principalDf = pd.DataFrame(data=data)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title('PCA explained variance',fontsize=20)
plt.savefig('reconstruction.png',bbox_inches='tight')
```

```python
# Computing the optimum number of clusters using the Elbow method

X=principalDf.iloc[:, range(1,51)].values #selecting the 50  PCA components

from sklearn.cluster import KMeans
wcss=[] #Within-Cluster-Sum-of-Squares method to determine the number of clusters

#this loop will fit the k-means algorithm to our data and 
#second we will compute the within cluster sum of squares and #appended to our wcss list.
for i in range(1,10):
    
    kmeans = KMeans(n_clusters=i, init ='k-means++', max_iter=300,  n_init=10,random_state=0 )
#i above is between 1-10 numbers. init parameter is the random initialization method  
#we select kmeans++ method. max_iter parameter the maximum number of iterations there can be to 
#find the final clusters when the K-means algorithm is running. we enter the default value of 300
#the next parameter is n_init which is the number of times the K_means algorithm will be run with
#different initial centroid.
    kmeans.fit(X)
    #kmeans algorithm fits to the X dataset
    wcss.append(kmeans.inertia_)
    #kmeans inertia_ attribute is:  Sum of squared distances of samples #to their closest cluster center.

#Plot the elbow graph
plt.plot(wcss)
plt.xlim((1,10))
plt.title('The Elbow Method Graph',fontsize=20)
plt.xlabel('Number of clusters',fontsize=20)
plt.ylabel('WCSS',fontsize=20)
plt.xticks(fontsize=14)
plt.savefig('Elbow.png',bbox_inches='tight')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init ='k-means++', max_iter=300, n_init=10,random_state=0 )
# We are going to use the fit predict method that returns for each #observation which cluster it belongs to. The cluster to which #client belongs and it will return this cluster numbers into a #single vector that is  called y K-means
y_kmeans = kmeans.fit_predict(X)
len(y_kmeans)
y_kmeans
```

#### Autoencoder based K-means clustering

```python
# change the kernel to python 3 for tensorflow to work smooth
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Dense, Input
import numpy as np

input_i = Input(shape=(166,))
encoded = Dense(units=50, activation='relu')(input_i)
decoded = Dense(units=166, activation='relu')(encoded)
autoencoder = Model(input_i,decoded)
```

```python
autoencoder.summary()
```

```python
autoencoder.compile(optimizer='adam', loss='mse')
```

```python
X_drug = np.loadtxt('X_drug.txt')

X_drug_name = np.loadtxt('X_drug_name.txt',delimiter='\n',dtype='str')
```

```python
autoencoder.fit(X_drug,X_drug, epochs=50, batch_size=1,shuffle=True)
```

```python
# Getting the hidden variables
from tensorflow.keras import backend as K
layerIndex = 1
func = K.function([autoencoder.get_layer(index=0).input], autoencoder.get_layer(index=layerIndex).output)
layerOutput = func([X_drug])  # input_data is a numpy array
layerOutput.shape
```
